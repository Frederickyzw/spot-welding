```python
import cv2
import matplotlib.pyplot as plt
from utils import *
from darknet import Darknet
```

```python
cfg_file = './yolov3_custom.cfg'
weight_file = './yolov3_custom_last.weights'
namesfile = './classes.names'
```

```python
m = Darknet(cfg_file)
m.load_weights(weight_file)
class_names = load_class_names(namesfile)
```

```python
plt.rcParams['figure.figsize'] = [24.0, 14.0]
```

```python
img = cv2.imread('./casting_512x512/def_front/cast_def_0_42.jpeg')
plt.imshow(img)
```

```python
original_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
resized_image = cv2.resize(original_image, (m.width, m.height))
```

```python
nms_thresh = 0.6
iou_thresh = 0.4
```

```python
boxes = detect_objects(m, resized_image, iou_thresh, nms_thresh)
```

```python
print_objects(boxes, class_names)
```

```python
plot_boxes(img, boxes, class_names, plot_labels=True, color = "Red")
```

```python
img = cv2.imread('./casting_512x512/ok_front/cast_ok_0_312.jpeg')
plt.imshow(img)

original_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
resized_image = cv2.resize(original_image, (m.width, m.height))

nms_thresh = 0.6
iou_thresh = 0.4

boxes = detect_objects(m, resized_image, iou_thresh, nms_thresh)
print_objects(boxes, class_names)
```
